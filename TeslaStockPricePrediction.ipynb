```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (16, 7)
```

```python
dataset = pd.read_csv('/content/TSLA.csv')
```

```python
# split 3 columns

X = dataset.iloc[:, 1:4].values  # 1, 2, 3
X
```

```python
Y = dataset.iloc[:, 4].values
Y
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

Y = Y.reshape(-1, 1)

X = sc.fit_transform(X)
Y = sc.fit_transform(Y)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=5)
```

```python
from sklearn.linear_model import LinearRegression

# instaintiate -> feed

lr_model = LinearRegression()
lr_model.fit(X_train, Y_train)
```

```python
lr_pred = lr_model.predict(X_test)
```

```python
from sklearn.metrics import r2_score

lr_accuracy = r2_score(Y_test, lr_pred)

lr_accuracy
```

```python
from sklearn.tree import DecisionTreeRegressor

# instainte -> feed

dtr_model =  DecisionTreeRegressor()

dtr_model.fit(X_train, Y_train)
```

```python
dtr_pred = dtr_model.predict(X_test)

dtr_accuracy = r2_score(Y_test, dtr_pred)

dtr_accuracy
```

```python
print(f"Linear Regresson: {lr_accuracy}")
print(f"Decision Tree Regressor: {dtr_accuracy}")
```
